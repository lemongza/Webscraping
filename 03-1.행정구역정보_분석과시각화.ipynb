```python
# data['행정구역'] == '경기도' 의 ['행정구역','인구수','면적']  컬럼의 데이터를 면적을 descending 순서대로 출력하기
data.loc[data['광역시도']=='경기도',['행정구역','인구수','면적']].sort_values(by='면적',ascending=True).reset_index(drop=True)
```

```python
#경기도에 속한 행정구역명, 인구수, 면적 선택하기 인구수를 descending 순서대로 출력하고 index reset 하여 gy_df 라는 변수에 저장하기
#data.loc[row , col]

gy_df = data.loc[data['광역시도']=='경기도',['행정구역','인구수','면적']].sort_values(by='인구수',ascending=False).reset_index(drop=True)
# 인구수2 라는 새로운 컬럼을 생성
gy_df['인구수2']=gy_df['인구수'].map('{:,}'.format)
gy_df['면적2']=gy_df['면적'].map('{:,}'.format)
gy_df
```

```python
# 면적이 > 1000 큰 조건과 매핑되는 새로운 DataFrame 생성 area_1000 라는 변수에 저장하기
area_100 = pd.DataFrame
area_100 = data.loc[data['면적']>1000,]
# area_1000 DataFrame 객체에 면적2 컬럼을 생성
area_100['면적2']=area_100['면적'].map('{:,}'.format)

area_100
```

```python
#경기도 인구수의 평균과 표준편차

print(f"경기도 인구수 평균 {(data.loc[data['광역시도']=='경기도', '인구수'].mean())}")
print(f"경기도 인구수 표준편차 {(data.loc[data['광역시도']=='경기도', '인구수'].std())}")
```

```python
#광역시도명을 인자로 받아서 해당 광역시에 속한 행정구역들의 인구수의 평균과 편차 구하기
def pop_mean_std(sido_name):
    print(f"{sido_name} 인구수 평균 = {(data.loc[data['광역시도']==sido_name, '인구수'].mean())}")
    print(f"{sido_name} 인구수 편차 = {(data.loc[data['광역시도']==sido_name, '인구수'].std())}")
```

```python
pop_mean_std('서울특별시')
```

```python
# Unique한 광역시도명 순회하면서  광역시도명을 출력합니다. 
for sido_name in data['광역시도'].unique():
    print(data,sido_name)
    
for i in data['광역시도'].unique():
    print(f"{i} 인구수 평균 = {(data.loc[data['광역시도']==i, '인구수'].mean())}")
    print(f"{i} 인구수 편차 = {(data.loc[data['광역시도']==i, '인구수'].std())}")
```

### 컬럼명 변경하기, 컬럼을 인덱스 전환하기
* rename() 함수
* inplace=False (default) 원본 DataFrame객체를 변경하지 않고, 변경한 결과만 출력한다.
* inPlace=True 원본 DataFrame객체를 변경하고, 변경한 결과는 출력하지 않는다.

```python
data.columns
```

```python
# 'Unnamed: 0' 컬럼명을 'seq' 로 변경한다. 원본 객체는 변경되지 않고 변경된 결과만 출력함
data.rename(columns={'Unnamed: 0':'seq'}) #inplace = False(default)
```

```python
#원본 객체가 변경되고 변경된 결과는 출력하지 않는다. (inplace=True)
data.rename(columns={'Unnamed: 0':'seq'},inplace=True)
data
```

#### 상관계수 (Correlation Coefficient)
